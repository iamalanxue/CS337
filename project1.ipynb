```python
gg = pd.read_json('gg2013.json', orient='records')
gg1 = pd.read_json('gg2015.json', orient='records')
```

```python
def getUser(x):
    return x['screen_name']
def getID(x):
    return x['id']
def removePunc(x):
#     new_string = x.translate(str.maketrans('', '', string.punctuation))
    return re.sub(r'[!?#.@:,\']', '', x)
gg['userid'] = gg['user'].apply(lambda x: getID(x))
gg['username'] = gg['user'].apply(lambda x: getUser(x))
gg['text'] = gg['text'].apply(lambda x: removePunc(str(x)))
gg['text'] = gg['text'].str.lower()
```

```python
gg.head()
```

```python
df = gg.loc[gg['text'].str.contains('hosts')]
```

```python
def nGram(text):
    test = text
    firstsplit = test.split(' hosts ', 1)
    if len(firstsplit) > 1:
        left = firstsplit[0].split(' ')
        left.reverse()
        right = firstsplit[1].split(' ')
        new_list = []
        x = ""
        for i in left:
            x =  i + x + ' '
            new_list.append(x)
        x = ""
        for i in right:
            x = x + i + ' '
            new_list.append(x)
        new_list = [x.rstrip() for x in new_list]
    else:
        new_list = []
    return new_list
```

```python
FINAL_NGRAMS = []
NGRAMS_DICT = {}
for i in range(len(df)):
    l = nGram(df['text'].iloc[i])
    for j in l:
        if j in NGRAMS_DICT:
            NGRAMS_DICT[j] += 1
        else:
            NGRAMS_DICT[j] = 1
```

```python
cd = sorted(NGRAMS_DICT.items(),key=operator.itemgetter(1),reverse=True)
```
